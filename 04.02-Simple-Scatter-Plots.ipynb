# 绘制简单的散点图

散点图与线图类似，只是点没有用线段连接起来:

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## 用 ``plt.plot``画散点图

想不到吧，散点图与线图用的是同一个方法，只是多了个参数``marker``, ``plt.plot``/``ax.plot`` 

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

以下是常用的marker

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

甚至将line码，color码和point吗合在一起用，如下所示:

```python
plt.plot(x, y, '-ok');
```

``plt.plot``支持更广泛的属性设置，如下:

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

## 用``plt.scatter``画散点图

这是更加强大的方法，与``plt.plot``类似:

```python
plt.scatter(x, y, marker='o');
```

``plt.scatter`` 与``plt.plot``最大的区别是可以控制每个点的颜色、尺寸灯，下图应为各点会覆盖，用alpha控制透明度 :

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

点的尺寸和颜色用来传达数据的意义，这样的方式很新颖.
例如下图坐标用叶片的长度和宽度来确定，点的大写用花的大下的100倍来表示，颜色又表示种类，这样一个图传达了丰富的含义

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

## ``plot`` Versus ``scatter``: 效率方面的提示

大的数据集倾向于用``plt.plot``
