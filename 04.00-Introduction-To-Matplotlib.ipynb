### ``show()`` or No ``show()``? How to Display Your Plots

三种方式可视化，脚本、IPython终端、IPython notebook(本文)


#### 脚本画图


``plt.show()`` 会启动事件循环，允许交互


```python
# ------- file: myplot.py ------
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```



```
$ python myplot.py
```
一个Python会话只能有一个show方法调用

#### IPython shell画图

在``ipython``里面启动 ``%matplotlib`` 命令，就进入Matplotlib模式，可以交互式画图，任何绘图命令就会更新图，但有的属性可能不更新，那就使用``plt.draw()``，没必要使用``plt.show()``

```ipython
In [1]: %matplotlib
Using matplotlib backend: TkAgg

In [2]: import matplotlib.pyplot as plt
```


#### 从notebook里面画图

- ``%matplotlib notebook`` 产生 *交互* 绘图
- ``%matplotlib inline`` will lead to *静态* 图片



````python
%matplotlib inline
````

只需要运行一次该命令

````python
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
````

## 两种风格的接口

传统的MATLAB风格, 更加强大的面向对象风格.

#### MATLAB风格

Matplotlib一开始目标用户是Matlab的用户
 MATLAB风格的工具都在pyplot (``plt``) interface

````python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
````

接口是有状态的，命令执行时是跟状态相关的，例如上面同样的plt.plot命令，产生的结果是不同的，可以说状态是全局的，而下面的风格，状态就记录在各个对象里面了

#### 面向对象的风格

画图函数都是``Figure``和``Axes``对象的显式的方法
To re-create the previous plot using this style of plotting, you might do the following:

````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
````
